```python
import numpy as np
import cv2
from matplotlib import pyplot as plt #비주얼 스튜디오 안에서 사진 보는 코드
```

```python

img = cv2.imread('mountain.jpg')
```

```python
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic') #비주얼 스튜디오 안에서 사진 보는 코드
plt.xticks([]), plt.yticks([]) 
plt.show()
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL) 
cv2.imshow('image',img)
cv2.waitKey(0)
k = cv2.waitKey(0)
if k == 27:
    cv2.destroyAllWindows()
elif k == ord('s'): #s 누르면 저장
    cv2.imwrite('mount.png',img)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture('tracking4.avi')
while(True):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'): #비디오 끄려면 q
        break
cap.release()
cv2.destroyAllWindows()
```

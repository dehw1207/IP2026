```python
import cv2
import numpy as np

def nothing(x):
    pass
#---------------------------------------
drawing = False # true if mouse is pressed

ix,iy = -1,-1

cx,cy = -1,-1

org_img = cv2.imread('mountain.jpg')
img = org_img.copy()

# mouse callback function
def draw_circle(event,x,y,flags,param):
    global ix,iy,cx,cy,drawing,img
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        cx,cy = x,y
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False


cv2.namedWindow('image')

cv2.createTrackbar('value','image',0,255,nothing)
cv2.setMouseCallback('image',draw_circle)
font = cv2.FONT_HERSHEY_SIMPLEX

while(1):
    img = org_img.copy()
    value = cv2.getTrackbarPos('value','image')
    
    if ix != -1 and iy != -1:
        sy = min(iy, cy)
        sx = min(ix, cx)
        ey = max(iy, cy)
        ex = max(ix, cx)
        img[sy:ey, sx:ex, 2] = value #sy,ey 좌표와 sx,ex 좌표 값의 안에 있는 블럭을 2(빨간색)으로 변경

    cv2.putText(img,f'Mouse position : ({ix},{iy})-({cx},{cy})-{value}', (10,30), font, 0.3,(255,255,255),1,cv2.LINE_AA)
    
    cv2.imshow('image',img)


    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
    
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('IMG_6504.jpg') #중간고사에 나옴 ㄷ 반대방향 순서로 좌표(시계방향)
rows,cols,ch = img.shape
pts1 = np.float32([[167,189],[762,193],[810,1096],[96,1083]])
pts2 = np.float32([[200,200],[800,200],[800,1100],[200,1100]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(900,1200))
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

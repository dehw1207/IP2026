#### For Loops

```python
for someChar in "Hello World":
    print (someChar)
```

```python
for (x, y) in [("a",1), ("b",2), ("c",3), ("d",4)]:
    print (x)
```

```python
for x in range(5):
    print (x)
```

```python
for i in range(len("mylist")):
    print (i, "mylist"[i])
```

```python
for (i, item) in enumerate("mylist"):
    print (i, item)
```

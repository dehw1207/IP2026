```python
li = [1, 11, 3, 4, 5]
```

```python
li.append('a')
li
```

```python
li.insert(2, 'i') # 중간고사가 이런식으로 코드가 주어지고 그거에 대한 insert 값을 적는 거 중요
li
```

```python
li.extend([9, 8, 7]) #append는 하나 씩, extend는 []를 통해서 여러 개 추가
li
```

```python
li.append([10, 11, 12]) #append는 대괄호를 사용하면 통째로 들어감
li
```

```python
li = ['a', 'b', 'c', 'b']

li.index('b') #위치
```

```python
li.count('b') #개수
```

```python
li.remove('b') #첫 번째 거 지우기
li
```

```python
li = [5, 2, 6, 8]
```

```python
li.reverse() #뒤집기
li
```

```python
li.sort() #오름차순 정렬
li
```

```python
tu = (2, 3, 5)

li = list(tu) #튜플인 tu를 리스트로 바꿔서 값 넣기
tu = tuple(li) #리스트인 li를 튜플로 바꿔서 값 넣기
```

```python
tu
```

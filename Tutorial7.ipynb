#### List Comprehensions

```python
li = [3, 6, 2, 7]
out = []
for elem in li:
    print(out)
    out.append(elem*2)
print(out)
```

```python
li = [3, 6, 2, 7]
[elem*12 for elem in li] #많이 쓰임
```

```python
li = [('a', 1), ('b', 2), ('c', 7)]
[ k * 3 for (x, k) in li]
```

```python
li = [3, 6, 2, 7, 1, 9]
[elem * 2 for elem in li if elem > 4]
```

```python
li = [3, 2, 4, 1]
[elem*2 for elem in 
    [item+1 for item in li] ] #뒤에부터 실행됨 (+1되고 *2)
```
